```python
def sepal_interaction(row: pd.Series) -> float:
  if row.sepal_length < 5:
    return row.sepal_length * 5 + row.sepal_width / 2

  if row.iris_type == 'Iris-virginica':
    return (row.sepal_width - row.petal_width) * 4

  return row.petal_width - 1

d['sepal_interaction'] = d.apply(sepal_interaction, axis='columns')
```

```python
p = pd.read_csv("/content/supermarket_prices.csv")
t = pd.read_csv("/content/supermarket_transactions.csv")
```

```python
p
```

```python
t
```

```python
t = t.join(other=p.set_index('Product'), on='Product', how='inner')
```

```python
t = t.rename(columns={'Price': 'Unit_Price'})
```

```python
t['Total_Price'] = t.Quantity * t.Unit_Price
```

```python
t = t.groupby(['Buyer', 'Product']).agg(
    {
        'Quantity': 'sum',
        'Unit_Price': 'max',
        'Total_Price': 'sum'
    }
).reset_index()
```
